# Corner Deviation Analysis

In this notebook, we will analyze the corner deviation data and visualize the results.

```python
from plot_utils import PlotUtils
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating Plots

First, we'll call the `draw_plots` method to create the plots and get the statistics.

```python
plot_paths = PlotUtils.draw_plots()
print(f"Created plots: {plot_paths}")
```

## Visualizing Plots

Now, let's display the created plots in the notebook.

```python
for path in plot_paths:
    plt.figure(figsize=(12, 8))
    img = plt.imread(path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(path)
    plt.show()
```

## Analyzing Results

Based on the generated plots and output statistics, we can draw the following conclusions:

1. The comparison plot of ground truth vs. predicted corners shows how accurately the model determines the number of corners in rooms.
2. The box plot of the deviations gives insight into the distribution of the deviations (mean, max and min).
3. The histogram of mean deviations allows us to assess which deviation values occur most frequently.

For a more detailed analysis, we can consider the following aspects:
- The percentage of rooms where the model accurately determined the number of corners.
- The mean corner deviation and its interpretation
- The presence of outliers in the data and their possible causes
